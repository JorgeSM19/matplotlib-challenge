```python
#Pongo mis dependencias
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#Pongo ruta de los archivos a utilizar
mouse_data = "Resources/mouse_drug_data.csv"
clinical_data = "Resources/clinicaltrial_data.csv"

#Leo los archivos a utilizar
mouse_df = pd.read_csv(mouse_data)
clinical_df = pd.read_csv(clinical_data)
mouse_df
```

```python
#Combino mis dos archivos en un solo Data Frame
data_merge = pd.merge(clinical_df, mouse_df, how="left", on=["Mouse ID", "Mouse ID"])
data_merge
```

# Tumor Response Treatment

```python
#Agrupo los tres valores a graficar y saco su promedio
tumor_data_mean = data_merge.groupby(["Drug", "Timepoint"])["Tumor Volume (mm3)"].mean()
tumor_data_mean_df = pd.DataFrame(tumor_data_mean)

tumor_data_mean_df = tumor_data_mean_df.reset_index()
```

```python
tumor_data_mean_df.head(10)
```

```python
#Saco el error estándar de los valores ya antes trabajados
tumor_value_sem = data_merge.groupby(["Drug", "Timepoint"])["Tumor Volume (mm3)"].sem()
tumor_value_sem_df = pd.DataFrame(tumor_value_sem)

tumor_value_sem_df = tumor_value_sem_df.reset_index()
```

```python
#Hago una tabla dinámica
tumor_mean_pivot = tumor_data_mean_df.pivot(index="Timepoint", columns="Drug")["Tumor Volume (mm3)"]
tumor_mean_pivot.head()
```

```python
tumor_sem_pivot = tumor_value_sem_df.pivot(index="Timepoint", columns="Drug")["Tumor Volume (mm3)"]
```

```python
plt.figure()
plt.errorbar(tumor_mean_pivot.index, tumor_mean_pivot["Capomulin"], yerr=tumor_sem_pivot["Capomulin"], color="r", fmt="o--")
plt.errorbar(tumor_mean_pivot.index, tumor_mean_pivot["Infubinol"], yerr=tumor_sem_pivot["Infubinol"], color="b", fmt="^--")
plt.errorbar(tumor_mean_pivot.index, tumor_mean_pivot["Ketapril"], yerr=tumor_sem_pivot["Ketapril"], color="g", fmt="s--")
plt.errorbar(tumor_mean_pivot.index, tumor_mean_pivot["Placebo"], yerr=tumor_sem_pivot["Placebo"], color="black", fmt="p--")

plt.title("Tumor Response to Treatment")
plt.ylabel("Tumor Volume (mm3)")
plt.xlabel("Time (Days)")
plt.legend(["Capomulin", "Infubinol", "Ketapril", "Placebo"], loc="best")
plt.grid()
plt.savefig("Images/Tumor_Response_Treatment")
plt.show()
```

# Metastatic Response to Treatment

```python
Metastatic_data_mean = data_merge.groupby(["Drug", "Timepoint"])["Metastatic Sites"].mean()
Metastatic_data_mean_df = pd.DataFrame(Metastatic_data_mean)

Metastatic_data_mean_df = Metastatic_data_mean_df.reset_index()
```

```python
Metastatic_data_sem = data_merge.groupby(["Drug", "Timepoint"])["Metastatic Sites"].sem()
Metastatic_data_sem_df = pd.DataFrame(Metastatic_data_sem)

Metastatic_data_sem_df = Metastatic_data_sem_df.reset_index()
```

```python
Meta_mean_pivot = Metastatic_data_mean_df.pivot(index="Timepoint", columns="Drug")["Metastatic Sites"]
Meta_mean_pivot.head()

Meta_sem_pivot = Metastatic_data_sem_df.pivot(index="Timepoint", columns="Drug")["Metastatic Sites"]
Meta_sem_pivot
```

```python
plt.figure()
plt.errorbar(Meta_mean_pivot.index, Meta_mean_pivot["Capomulin"], yerr=Meta_sem_pivot["Capomulin"], color="r", fmt="o--")
plt.errorbar(Meta_mean_pivot.index, Meta_mean_pivot["Infubinol"], yerr=Meta_sem_pivot["Infubinol"], color="b", fmt="^--")
plt.errorbar(Meta_mean_pivot.index, Meta_mean_pivot["Ketapril"], yerr=Meta_sem_pivot["Ketapril"], color="g", fmt="s--")
plt.errorbar(Meta_mean_pivot.index, Meta_mean_pivot["Placebo"], yerr=Meta_sem_pivot["Placebo"], color="black", fmt="p--")

plt.title("Metastatic Spread During Treatment")
plt.ylabel("Met. Sites")
plt.xlabel("Time (Days)")
plt.legend(["Capomulin", "Infubinol", "Ketapril", "Placebo"], loc="best")
plt.grid()
plt.savefig("Images/Metastatic Response to Treatment")
plt.show()
```
